```python
# def snippet_transformer(block):
#     lines = block.split('# ')
#     results = []
#     for line in lines:
#         if line.startswith('  '):
#             results.append(f'"    {line.strip()}",')
#         else:
#             results.append(f'"{line.strip()}",')

#     return results

def snippet_transformer(block):
    lines = block.split('# ')[1:]
    results = []
    
    for line in lines:
        leading_space = len(line) - len(line.lstrip())
        space = leading_space * ' '
        results.append(f'"{space}{line.strip()}",')
    
    return results

def print_snippet(title, body, prefix, description):
    results = snippet_transformer(body)

    print(f'"{title}": {{')
    print(f'  "prefix": "{prefix}",')
    print(f'  "body": [')
    for line in results: 
        print(f'    {line}')

    print(f'  ],')
    print(f'  "description": "{description}"')
    print(f'}}')
    
    return None
```
